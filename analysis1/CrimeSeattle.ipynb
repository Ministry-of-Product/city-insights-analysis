```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc = {'figure.figsize': (10, 8)})

crime_data = pd.read_csv('../data/Crime_Data.csv')
crime_data.head()
```

```python
crime_data.dtypes
```

```python
# Check for NaN values in the "MCPP" column and fill them with an empty string if necessary.
crime_data['MCPP'].fillna('', inplace=True)

# Filter the dataset to include only rows where the "MCPP" column contains the word "ballard."
ballard_data = crime_data[crime_data['MCPP'].str.contains('ballard', case=False, na=False)]
```

```python
ballard_data.shape
```

```python
# # Assuming your datetime string is stored in a variable called 'datetime_str'
# datetime_str = "2020-02-05 00:51:21"

# # Convert the datetime string to a datetime object
# datetime_obj = pd.to_datetime(datetime_str)
```

```python
ballard_data['Offense Start DateTime'] = pd.to_datetime(ballard_data['Offense Start DateTime'])
```

```python
ballard_data.dtypes
```

```python
from datetime import datetime, timedelta

# Calculate the date one month ago from the current date
one_month_ago = datetime.now() - timedelta(days=30)
```

```python
ballard_last_month_data = ballard_data[ballard_data['Offense Start DateTime'] >= one_month_ago]
```

```python
ballard_last_month_data.shape
```

```python
ballard_last_month_data
```

```python
# Get a list of all unique "Offense Parent Group" types
offense_parent_group_types = ballard_last_month_data['Offense Parent Group'].unique()

# Convert the result to a Python list if needed
offense_parent_group_types_list = list(offense_parent_group_types)
```

```python
offense_parent_group_types_list
```

```python
# Count the number of unique "Offense Parent Group" types
num_offense_parent_group_types = ballard_last_month_data['Offense Parent Group'].nunique()
num_offense_parent_group_types
```

```python
# Count occurrences of each "Offense Parent Group" and store the result in a new DataFrame
offense_parent_group_counts = ballard_last_month_data['Offense Parent Group'].value_counts().reset_index()

# Rename the columns for clarity
offense_parent_group_counts.columns = ['Offense Parent Group', 'Count']
offense_parent_group_counts
```

```python
# Filter the data to include only Ballard-specific rows and "ASSAULT OFFENSES" with the specified time frame
ballard_assault_last_month_data = ballard_data[
    (ballard_data['MCPP'].str.contains('ballard', case=False)) &
    (ballard_data['Offense Parent Group'] == 'ASSAULT OFFENSES') &
    (ballard_data['Offense Start DateTime'] >= one_month_ago)
]
ballard_assault_last_month_data
```
