```python
import pandas as pd
import folium

# Read the data
data = pd.read_csv('../data/clean_crime_data.csv')

# Initialize the map centered around the average latitude and longitude
m = folium.Map(location=[data['latitude'].mean(), data['longitude'].mean()], zoom_start=12)

# Add a marker for each crime incident
for _, row in data.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=f"<b>Date:</b> {row['Date']}<br><b>Location:</b> {row['Approx Loc']}<br><b>Description:</b> {row['Description']}"
    ).add_to(m)

# Save the map as an HTML file
m.save('crime_map.html')

# Display the map in the Jupyter Notebook
m
```
